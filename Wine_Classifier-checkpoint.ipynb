<h1 style="font-weight: bold">Introduction</h1>

In this project we intend to use a dataset obtained from a thorough chemical examination of these wines in order to use machine learning to identify wines from various varietals grown in the same region of Italy.

There are 178 examples in the collection, each of which represents a unique wine sample with 13 chemical components, such as alcohol percentage, acidity, phenolic compounds, and color intensity. The wine cultivar, which is divided into three different groups, is the target variable for this classification task. 

 We will use Random Forest, k-Nearest Neighbors (k-NN), Support Vector Machines (SVM), and Logistic Regression to test the dataset. We want to determine the best method for precisely forecasting the wine cultivar based on its chemical characteristics by creating a baseline model and contrasting the results of various algorithms.


<h1 style="font-weight: bold">Importing libraries</h1>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, label_binarize
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix, roc_auc_score, roc_curve, auc
```

<h1 style="font-weight: bold">Loading Data</h1>

```python
wine_data = pd.read_csv('C://Users//smuka//Desktop//ML Project//wine//wine_data.csv')
```

```python
#The statistical characteristics of the numerical features in the dataset.

wine_data.describe
```

<h1 style="font-weight: bold">Data Preprocessing: Splitting and Scaling</h1>

```python
# Split features and target
X = wine_data.drop('Class', axis=1)
y = wine_data['Class']
```

```python
# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

<h1 style="font-weight: bold">Applying PCA (2 components)</h1>

```python
# PCA for visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=y, palette='Set1', s=70)
plt.title('PCA of Wine Dataset (2D)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Class')
plt.grid(True)
plt.tight_layout()
plt.show()
```

<h1 style="font-weight: bold">Setup for Training and Evaluating ML Models</h1>

```python
# Prepare metric holders
model_names = []
accuracies = []
f1_scores = []
roc_aucs = []
```

```python
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y)
```

```python
# Initialize models
models = {
    'Logistic Regression': LogisticRegression(max_iter=1000),
    'SVM': SVC(probability=True),
    'KNN': KNeighborsClassifier(),
    'Random Forest': RandomForestClassifier()
}

# Result dictionary
results = {}
```

<h1 style="font-weight: bold">Model Training, Evaluation, and Results Compilation</h1>

```python
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    y_score = model.predict_proba(X_test)

    acc = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred, average='weighted')
    cm = confusion_matrix(y_test, y_pred)

    # Binarize labels for ROC AUC
    y_bin = label_binarize(y_test, classes=[1, 2, 3])
    auc_score = roc_auc_score(y_bin, y_score, multi_class='ovr')

    results[name] = {
        'Accuracy': acc,
        'F1 Score': f1,
        'Confusion Matrix': cm,
        'ROC AUC': auc_score,
        'y_score': y_score,
        'y_bin': y_bin
    }

   # Save for bar plots
    model_names.append(name)
    accuracies.append(acc)
    f1_scores.append(f1)
    roc_aucs.append(auc_score)
```

```python
# Print results
for name, metrics in results.items():
    print(f"\n{name}")
    print(f"Accuracy: {metrics['Accuracy']:.4f}")
    print(f"F1 Score: {metrics['F1 Score']:.4f}")
    print(f"ROC AUC: {metrics['ROC AUC']:.4f}")
    print("Confusion Matrix:")
    print(metrics['Confusion Matrix'])
```

<h1 style="font-weight: bold">Graphical representation of the evaluations</h1>

<h2 style="font-weight: bold">Confusion matrices</h2>

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes = axes.flatten()

for idx, (name, metrics) in enumerate(results.items()):
    sns.heatmap(metrics['Confusion Matrix'], annot=True, fmt='d', cmap='Blues', ax=axes[idx], cbar=False)
    axes[idx].set_title(f'{name} - Confusion Matrix')
    axes[idx].set_xlabel('Predicted')
    axes[idx].set_ylabel('True')

plt.tight_layout()
plt.show()
```

<h2 style="font-weight: bold">ROC curves</h2>

```python
# Plot ROC curves
colors = ['red', 'green', 'blue']
classes = [1, 2, 3]

for name, metrics in results.items():
    y_bin = metrics['y_bin']
    y_score = metrics['y_score']

    fpr = dict()
    tpr = dict()
    roc_auc = dict()

    for i in range(3):  # 3 classes
        fpr[i], tpr[i], _ = roc_curve(y_bin[:, i], y_score[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])

    # Plot all ROC curves for this model
    plt.figure(figsize=(8, 6))
    for i in range(3):
        plt.plot(fpr[i], tpr[i], color=colors[i],
                 label=f'Class {classes[i]} (AUC = {roc_auc[i]:.2f})')

    plt.plot([0, 1], [0, 1], 'k--', lw=1)
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.title(f'ROC Curves - {name}')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.legend(loc='lower right')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

<h2 style="font-weight: bold">Bar charts of performance metrics</h2>

```python
plt.figure(figsize=(14, 4))

# Accuracy
plt.subplot(1, 3, 1)
plt.barh(model_names, accuracies, color='skyblue')
plt.title('Model Accuracy')
plt.xlabel('Accuracy')

# F1 Score
plt.subplot(1, 3, 2)
plt.barh(model_names, f1_scores, color='orange')
plt.title('Model F1 Score')
plt.xlabel('F1 Score')

# ROC AUC
plt.subplot(1, 3, 3)
plt.barh(model_names, roc_aucs, color='green')
plt.title('Model ROC AUC')
plt.xlabel('ROC AUC')

plt.tight_layout()
plt.show()
```

<h2 style="font-weight: bold">Feature importances accross models</h2>

```python
# Fit models
logreg = LogisticRegression(max_iter=1000)
logreg.fit(X_train, y_train)

rf = RandomForestClassifier()
rf.fit(X_train, y_train)

svm = SVC(kernel='linear', probability=True)
svm.fit(X_train, y_train)
```

```python

# Get feature names
feature_names = X.columns

# Random Forest Importance
rf_importances = rf.feature_importances_

# Logistic Regression Importance (average across classes)
logreg_importances = np.mean(np.abs(logreg.coef_), axis=0)

# SVM Importance (average across classes)
svm_importances = np.mean(np.abs(svm.coef_), axis=0)

# Combine into a DataFrame for plotting
importance_df = pd.DataFrame({
    'Feature': feature_names,
    'Random Forest': rf_importances,
    'Logistic Regression': logreg_importances,
    'SVM (Linear)': svm_importances
})

# Melt the DataFrame for seaborn plotting
importance_melted = importance_df.melt(id_vars='Feature', 
                                       var_name='Model', 
                                       value_name='Importance')

# Plot
plt.figure(figsize=(12, 8))
sns.barplot(data=importance_melted, x='Importance', y='Feature', hue='Model')
plt.title("Feature Importance Across Models")
plt.tight_layout()
plt.show()
```
